```python
# Write you web scraping code here.
import pandas as pd
```

```python
url = "https://www.basketball-reference.com/leagues/NBA_2021_per_game.html"
```

```python
bbdf = pd.read_html(url, header=0)
bbdf[0]
```

```python
bbdf_1 = bbdf[0]
bbdf_1
```

```python
bbdf_1[bbdf_1.Age == 'Age']
```

```python
bbdf_2 = bbdf_1.drop(bbdf_1[bbdf_1.Age == 'Age'].index)
bbdf_2
```

```python
bbdf_2.isnull().sum().sum()
```

```python
bbdf_3 = bbdf_2.fillna(0.0)
bbdf_3
```

```python
bbdf2021 = bbdf_3.drop(bbdf_3[bbdf_3.Tm == 'TOT'].index)
bbdf2021
```

```python
bbdf2021.to_csv('bb_data.csv', index=False)
```
